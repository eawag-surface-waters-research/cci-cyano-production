```python
import os
import netCDF4
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
file = "/home/runnalja/git/cci-cyano-production/data/327.nc"
```

Timeseries at one pixel

```python
with netCDF4.Dataset(file) as nc:
    x = nc.variables["chla"][:, 30, 50]
    t = nc.variables["time"]
    time = [datetime.fromtimestamp(int(ts)) for ts in np.array(t)]
    plt.plot(time, x)
    plt.show()
```

At specific time point

```python
with netCDF4.Dataset(file) as nc:
    x = nc.variables["chla"][0, :, :]
    plt.imshow(x)
    plt.colorbar()
    plt.show()
```

Number of values in dataset

```python
with netCDF4.Dataset(file) as nc:
    x = nc.variables["summary"][:, :]
    plt.imshow(x)
    plt.colorbar()
    plt.show()
```
